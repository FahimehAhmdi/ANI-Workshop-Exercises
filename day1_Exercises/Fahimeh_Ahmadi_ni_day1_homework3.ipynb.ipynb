```python
import os

new_setting = "* annex.largefiles=(mimeencoding=binary)\n"
if os.path.exists(".gitattributes"):
    with open(".gitattributes", "r", encoding="utf-8") as f:
        lines = f.readlines()

    if lines:
        lines[-1] = new_setting
    else:
        lines = [new_setting]
else:
    lines = [new_setting]

with open(".gitattributes", "w", encoding="utf-8") as f:
    f.writelines(lines)
```

```python
new_condition = "* annex.largefiles=((mimeencoding=binary)or(largerthan=5kb))\n"

if os.path.exists(".gitattributes"):
    with open(".gitattributes", "r", encoding="utf-8") as f:
        lines = f.readlines()

    if lines:
        lines[-1] = new_condition
    else:
        lines = [new_condition]
else:
    lines = [new_condition]

with open(".gitattributes", "w", encoding="utf-8") as f:
    f.writelines(lines)
```
